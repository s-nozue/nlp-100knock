### 64. アナロジーデータでの実験
単語アナロジーの評価データをダウンロードし，vec(2列目の単語) - vec(1列目の単語) + vec(3列目の単語)を計算し，そのベクトルと類似度が最も高い単語と，その類似度を求めよ．求めた単語と類似度は，各事例の末尾に追記せよ．

```python
with open('data/questions-words.txt') as fi, open('work/questions-words-ans.txt', 'w') as fo:
    for l in fi:
        l = l.strip().split(" ")
        if l[0] == ":":
            fo.write(" ".join(l) + "\n")
        else:
            most_sim = word_vectors.most_similar(positive=[l[1], l[2]], negative=[l[0]], topn=1)
            fo.write(" ".join(l) + " " + most_sim[0][0] + " " + str(most_sim[0][1]) + "\n")
```

### 65. アナロジータスクでの正解率
64の実行結果を用い，意味的アナロジー（semantic analogy）と文法的アナロジー（syntactic analogy）の正解率を測定せよ．

```python
with open('work/questions-words-ans.txt') as f:
    for l in f:
        if l.startswith(':'):
            print(l, end='')
```

```python
from sklearn.metrics import accuracy_score

golds = []
preds = []

with open('work/questions-words-ans.txt') as f:
    for l in f:
        l = l.strip().split(" ")
        if l[0] == ":":
            if l[1] == 'gram1-adjective-to-adverb':
                semantic_accuracy = accuracy_score(golds, preds)
                golds = []
                preds = []
        else:
            golds.append(l[3])
            preds.append(l[4])

syntactic_accuracy = accuracy_score(golds, preds)

print("semantic_accuracy:", semantic_accuracy)
print("syntactic_accuracy:", syntactic_accuracy)
```

```python
import pandas as pd
from scipy.stats import spearmanr
df = pd.read_csv('data/combined.csv')

df['similarities'] = [word_vectors.similarity(w1, w2) for w1, w2 in zip(df['Word 1'].values, df['Word 2'].values)]
df['Human (rank)'] = df['Human (mean)'].rank(ascending=False, method='min')  # ascending=False: 降順
df['similarities (rank)'] = df['similarities'].rank(ascending=False, method='min')

correlation, pvalue = spearmanr(df['Human (rank)'], df['similarities (rank)'])
print("correlation:", correlation)
print("pvalue:", pvalue)
df.head()
```

### 67. k-meansクラスタリング
国名に関する単語ベクトルを抽出し，k-meansクラスタリングをクラスタ数k=5として実行せよ

```python
country_words = set()
with open('data/questions-words.txt') as f:
    for l in f:
        l = l.rstrip().split(" ")
        if l[1] == 'currency':
            break
        elif l[0] == ":":
            continue
        else:
            country_words.add(l[1])
country_words = list(country_words)
print(country_words)
country_vectors = [word_vectors[w] for w in country_words]
len(country_vectors)
```

```python
from sklearn.cluster import KMeans # K-means クラスタリングをおこなう
# この例では 3 つのグループに分割 (メルセンヌツイスターの乱数の種を 10 とする)
kmeans_model = KMeans(n_clusters=5, random_state=10, n_init='auto', init='random').fit(country_vectors)
# 分類結果のラベルを取得する
labels = kmeans_model.labels_
# 分類結果を確認する
print(labels)
```

```python
result = {}
for label, word in zip(labels, country_words):
    if label not in result:
        result[label] = []
    result[label].append(word)
result = dict(sorted(result.items(), key=lambda x: x[0]))   # 下のfor文でタプルを扱うのは面倒くさいのでdictに戻す
for label, words in result.items():
    print(f'label = {label}')
    print(words)
```

### 68. Ward法によるクラスタリング
国名に関する単語ベクトルに対し，Ward法による階層型クラスタリングを実行せよ．さらに，クラスタリング結果をデンドログラムとして可視化せよ．

```python
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import linkage, dendrogram
 
linkage_result = linkage(country_vectors, method='ward')
plt.figure(num=None, figsize=(16, 9), dpi=200, facecolor='w', edgecolor='k')
dendrogram(linkage_result, labels=country_words, leaf_font_size=8.)
plt.show()
```

### 69. t-SNEによる可視化
ベクトル空間上の国名に関する単語ベクトルをt-SNEで可視化せよ．

```python
from sklearn.manifold import TSNE
import numpy as np
import matplotlib.pyplot as plt


country_vectors = np.array(country_vectors)

# t-SNE で次元圧縮する
tsne = TSNE(n_components=2, random_state=42)
digits_tsne = tsne.fit_transform(country_vectors)

# クラス毎に色分けしつつ、各単語の名前もプロットする
plt.figure(figsize=(20, 20))
for word, x, y in zip(list(country_words), digits_tsne[:, 0], digits_tsne[:, 1]):
    if word in result[0]:
        plt.scatter(x, y, c='blue')
    elif word in result[1]:
        plt.scatter(x, y, c='red')
    elif word in result[2]:
        plt.scatter(x, y, c='green')
    elif word in result[3]:
        plt.scatter(x, y, c='yellow')
    elif word in result[4]:
        plt.scatter(x, y, c='purple')
    plt.annotate(word, xy=(x, y), xytext=(0, 0), textcoords='offset points')
plt.show()
```
